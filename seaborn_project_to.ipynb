```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from collections import Counter
import warnings
warnings.filterwarnings('ignore') 
```

```python
df = pd.read_csv("PoliceKillingsUS.csv", encoding= "windows-1252")
df.head()
```

```python
df.info()
```

```python
df.shape
```

Box plot of age in both genders separeted by manner of death.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='manner_of_death', y='age', hue='gender', data=df)
plt.xlabel('Manner of Death')
plt.ylabel('Age')
plt.legend()
plt.show()
```

Demonstrate the Qurtiles of age in both genders separeted by manner of death.

```python
plt.figure(figsize=(6, 4))

sns.boxplot(data=df, x='gender', y='age', hue='manner_of_death')


plt.legend(loc='upper right' )

# Grafiği çizdiriyorum
plt.show()
```

 Demonstrate the age in both genders separeted by manner of death by a swarm plot.

```python
plt.figure(figsize=(5, 3))
sns.swarmplot(data=df, x='gender', y='age', hue='manner_of_death')

plt.legend()
plt.show()
```

12 most dangerous cities

```python
plt.figure(figsize=(7, 4))
sns.barplot(x=df['city'].value_counts().head(12).index, y= df['city'].value_counts().head(12).values, palette='rocket')

plt.title('Most Dangerous Cities')


plt.xticks(rotation=45, size = 8)  

plt.show()
```

Plot 20 most dangerous states

```python
plt.figure(figsize=(7, 4))
sns.barplot(x=df['state'].value_counts().head(20).index, 
            y=df['state'].value_counts().head(20).values, 
            palette='coolwarm')

plt.title('20 Most Dangerous States')

plt.xticks(rotation = 60, size = 7)

plt.show()
```

Plot Having mental ilness or not for killed people"

```python
plt.figure(figsize=(7, 5))
sns.barplot(x=df['signs_of_mental_illness'].value_counts().index, 
            y=df['signs_of_mental_illness'].value_counts().values, 
            palette='rocket')

plt.xticks(size = 10)
plt.show()
```

Plot number of Threat Types

```python
plt.figure(figsize=(5, 4))
sns.barplot(x = df['threat_level'].value_counts().index, 
            y = df['threat_level'].value_counts().values, 
            palette='rocket')

# Grafiği çizdiriyorum
plt.show()
```
